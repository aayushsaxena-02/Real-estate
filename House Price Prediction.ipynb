# train test split using scikit learn

```python
from sklearn.model_selection import train_test_split
train_set,test_set = train_test_split(housing,test_size=0.2,random_state=42)
```

```python
#stratified Sampling
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits = 1,test_size=0.2,random_state=42)
for train_index, test_index in split.split(housing,housing['CHAS']):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
```

```python
housing = strat_train_set
housing.shape
```

```python
#target values
housing = strat_train_set.drop("MEDV",axis=1)
housing_labels = strat_train_set["MEDV"].copy()
```

```python
# Fill missing entries
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy="median")
imputer.fit(housing)
```

```python
# median of each feature and will replace with this values if and NA
imputer.statistics_
```

```python
housing.head()
```

```python
X = imputer.transform(housing)
```

```python
housing_tr = pd.DataFrame(X,columns=housing.columns)
```

```python
housing_tr.describe()
```

# Creating a Pipeline


```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

my_pipeline = Pipeline([
    ('imputer',SimpleImputer(strategy = "median")),
    ('std_scalar',StandardScaler()),
])
```

```python
# it is a numpy array & predictor take array as input 
housing_num_tr = my_pipeline.fit_transform(housing_tr)
```

```python
housing.head()
```

```python
housing_num_tr.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
# model = LinearRegression()
# model = DecisionTreeRegressor()
model = RandomForestRegressor()
model.fit(housing_num_tr,housing_labels)
```

# Predicting the model

```python
# Testing with few data
some_data = housing.iloc[:5]
some_labels = housing_labels.iloc[:5]
prepared_data = my_pipeline.transform(some_data)
model.predict(prepared_data)
# will give array of predictions
```

```python
list(some_labels)
#original values
```

# Evaluating the model

```python
import numpy as np
from sklearn.metrics import mean_squared_error
housing_predictions = model.predict(housing_num_tr)
mse = mean_squared_error(housing_labels,housing_predictions)
rmse = np.sqrt(mse)
```

```python
# model undergone overfitting
rmse
```

# Better model evaluation technique - cross validation 

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model,housing_num_tr,housing_labels,scoring="neg_mean_squared_error",cv=10)
rmse = np.sqrt(-scores)
```

```python
rmse
```

```python
def print_scores(scores):
    print("Scores are: ",scores)
    print("Mean is: ",scores.mean())
    print("Std is: ",scores.std())
```

```python
print_scores(rmse)
```

# Saving the model

# Testing the model

```python
X_test = strat_test_set.drop("MEDV",axis=1)
Y_test = strat_test_set["MEDV"].copy()
X_test_prepared = my_pipeline.transform(X_test)
final_predictions = model.predict(X_test_prepared)
final_mse = mean_squared_error(Y_test,final_predictions)
final_rmse = np.sqrt(final_mse)
```

```python
final_rmse
```

```python
from sklearn.metrics import r2_score
accuracy = r2_score(Y_test,final_predictions)
print("Accuracy of model is:",accuracy*100,'%')
```
